```python
import pandas as pd
import sqlite3


df = pd.read_csv('pancreatic_cancer_prediction_sample.csv', nrows=10000)
display(df.head())
df.info()

#Transform the data
#dropping duplicates 
df = df.drop_duplicates()

#fill missing values with mean
df.fillna(df.mean(numeric_only = True), inplace = True)

#fill missing categorical values with "unknown"
df.fillna("Unknown", inplace = True)

display(df.head())
print("Let's check if there are any missing values.....")
print(df.isnull().sum())

#Load the data

conn = sqlite3.connect('pancreatic_cancer_clean.db')
cursor = conn.cursor()
cursor.execute("""
CREATE TABLE IF NOT EXISTS pancreatic_cancer_clean(
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    Age INTEGER,
    Gender TEXT,
    SMOKING_HISTORY INTEGER,
    OBESITY INTEGER,
    DIABETES INTEGER,
    Chronic_Pancreatitis INTEGER,
    Family_History INTEGER,
    Hereditary_Condition INTEGER,
    Back_Pain INTEGER,
    Jaundice INTEGER,
    Abdominal_Discomfort  INTEGER,
    Weight_Loss   INTEGER,  
    Development_of_Type2_Diabetes  INTEGER, 
    Stage_at_Diagnosis TEXT,
    Survival_Time_Months INTEGER,  
    Treatment_Type TEXT,
    Survival_Status INTEGER,
    Alcohol_Consumption INTEGER,
    Physical_Activity_Level INTEGER,
    Diet_Processed_Food TEXT,
    Access_to_Healthcare TEXT,
    Urban_vs_Rural TEXT,
    Economic_Status TEXT  
);
""")
conn.commit()

#Insert cleaned  data into SQL Lite
df.to_sql('pancreatic_cancer_clean', conn, if_exists='replace', index=False)

#query to validate the pipeline
query = "select  * from pancreatic_cancer_clean order by age desc limit 5;"
result = pd.read_sql(query, conn)
display(result)

#close connection
conn.close()
```
